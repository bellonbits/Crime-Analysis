```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('crime.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
```

```python
missing_values[missing_values > 0]
```

```python
# Specify the date format if your data follows a standard format like 'YYYY-MM-DD'
data['Date Rptd'] = pd.to_datetime(data['Date Rptd'], format='%Y-%m-%d', errors='coerce')
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], format='%Y-%m-%d', errors='coerce')
```

```python
data.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Crm Cd Desc', data=data, order=data['Crm Cd Desc'].value_counts().index)
plt.title('Distribution of Crime Types')
plt.xlabel('Number of Incidents')
plt.ylabel('Crime Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Vict Age'], bins=30, kde=True)
plt.title('Distribution of Victim Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Vict Sex', hue='Crm Cd Desc')
plt.title('Crime Type by Victim Sex')
plt.xlabel('Victim Sex')
plt.ylabel('Count')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Extract year and month for analysis
data['Year'] = data['DATE OCC'].dt.year
data['Month'] = data['DATE OCC'].dt.month

monthly_crime = data.groupby(['Year', 'Month']).size().reset_index(name='Counts')

plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_crime, x='Month', y='Counts', hue='Year', marker='o')
plt.title('Monthly Crime Trends Over Years')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.xticks(range(1, 13))
plt.show()
```

```python
import folium

# Create a base map
m = folium.Map(location=[data['LAT'].mean(), data['LON'].mean()], zoom_start=12)

# Add points for each crime
for index, row in data.iterrows():
    folium.CircleMarker(
        location=(row['LAT'], row['LON']),
        radius=5,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.5
    ).add_to(m)

# Save the map
m.save('crime_map.html')
from IPython.display import Image, display

# Replace 'path_to_image.jpg' with the actual path to your image
image_path = '1.png'
display(Image(filename=image_path))
```
